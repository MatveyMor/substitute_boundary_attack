```python
result = torch.load('save_cifar/dct_targeted_ResNet18_1000_0_32_0.2000_rand.pth')
res_image = torch.load('save_cifar/images_ResNet18_1000.pth')
```

```python
result['adv'][0].shape
```

```python
input_or_adv = result['adv'][0].to('cpu')
starting_points = res_image['images'][0].to('cpu')
```

```python

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=[10, 20])

plt.subplot(121)
plt.imshow(np.array(input_or_adv).transpose((1,2,0)))
plt.title('Adversarial image')

plt.subplot(122)
plt.imshow(np.array(starting_points).transpose((1,2,0)))
plt.title('init image')

plt.show()
```

```python
dstncs = []
for i in range(result['adv'].shape[0]):
    input_or_adv = result['adv'][i].to('cpu')
    starting_points = res_image['images'][i].to('cpu')
    distance = np.linalg.norm(input_or_adv.cpu().detach().numpy() - starting_points.cpu().detach().numpy())
    dstncs.append(distance / (3*32*32))
```
